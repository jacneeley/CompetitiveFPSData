```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import csv
```

```python
url = 'https://prosettings.net/cs-go-pro-settings-gear-list/'
page = requests.get(url)
page
```

```python
soup = bs(page.content,"html.parser")
soup.title.text
```

```python
#code to grab data from tables
table = soup.find('table',{'id':'table_1'})
headers =[]
for i in table.find_all('th'):
    col = i.text.strip()
    headers.append(col)
headers
```

```python
#create dataframe with headers from table
df = pd.DataFrame(columns = headers)
df
```

```python
#loop through each row and grab td tags from table to populate each row in the df
for row in table.find_all('tr')[1:]: #exclude header
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
df = df[:510]
df
```
