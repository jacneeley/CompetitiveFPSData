```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
prodata = pd.read_csv('prodata_refined.csv')
prodata
```

```python
monitors = prodata.groupby(by=prodata['Monitor']) #Count of monitor used for each game
monitors2 = monitors['Game'].count().plot(kind = 'bar', title = 'Monitors used by CSGO and Valorant Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Brand',figsize=(12, 6))
```

```python
val = prodata[prodata['Game']== 'VALORANT']#Visualising all monitors used by Valorant players
monitors3 = val['Monitor'].value_counts().plot(kind = 'bar', title = 'Monitors used by Valorant Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Count',figsize=(14, 6)) 
```

```python
csgo = prodata[prodata['Game']== 'Counter Strike: Global Offensive'] #Visualising all monitors used by CSGO players
monitors4 = csgo['Monitor'].value_counts().plot(kind = 'bar', title = 'Monitors used by CSGO Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Count',figsize=(14, 6)) 
```

```python
mouse = prodata.groupby(['Mouse']) #Obtaining most popular mouse both games
m = pd.DataFrame(mouse['Mouse'].count())
m.rename(columns = {'Mouse': 'In Use'}, inplace = True)
m[m['In Use'] == m['In Use'].max()]
```

```python
csgo2 = prodata[prodata['Game']== 'Counter Strike: Global Offensive'] #Visualising all mouses used by CSGO players
mouse2 = csgo2['Mouse'].value_counts().plot(kind = 'bar', title = 'Mouse used by CSGO Players',
              xlabel ='Mouse Brand',ylabel = 'Mouse Count',figsize=(17, 6)) 
```

```python
val2 = prodata[prodata['Game']== 'VALORANT']#Visualising all mouses used by Valorant players
mouse2 = val2['Mouse'].value_counts().plot(kind = 'bar', title = 'Mouse used by Valorant Players',
              xlabel ='Mouse Brand',ylabel = 'Mouse Count',figsize=(14, 6)) 
```

```python
GPU = prodata.groupby(['GPU']) #Obtaining most popular graphics card both game
x = pd.DataFrame(GPU['GPU'].count())
x.rename(columns = {'GPU': 'In Use'}, inplace = True)
x[x['In Use'] == x['In Use'].max()]
```
