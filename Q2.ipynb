```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
prodata = pd.read_csv('prodata_refined.csv')
prodata
```

```python
monitors = prodata.groupby(by=prodata['Monitor']) #Count of monitor used for each game
monitors['Game'].count()
```

```python
sns.displot(x=prodata['Monitor'], bins=40)
```

```python
heights = [50, 100, 150, 200, 250]

plt.bar(monitors['Game'].count(), heights, width= 2.5, color = 'g')
plt.title('Barchart of Monitors Used')
plt.ylabel('Monitor Brand')
plt.xlabel('Count')

plt.show()
```
