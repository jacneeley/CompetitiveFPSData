```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
prodata = pd.read_csv('prodata_refined.csv')
prodata
```

```python
monitors = prodata.groupby(by=prodata['Monitor']) #Count of monitor used for each game
monitors2 = monitors['Game'].value_counts()
```

```python
monitors2.plot(kind = 'bar', title = 'Monitors used by CSGO and Valorant Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Brand',figsize=(14, 6)) 
```

```python
val = prodata[prodata['Game']== 'VALORANT']
monitors3 = val['Monitor'].value_counts()
monitors3.plot(kind = 'bar', title = 'Monitors used by Valorant Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Brand',figsize=(14, 6)) 
```

```python
csgo = prodata[prodata['Game']== 'Counter Strike: Global Offensive']
monitors4 = csgo['Monitor'].value_counts()
monitors4.plot(kind = 'bar', title = 'Monitors used by CSGO Players',
              xlabel ='Monitor Brand',ylabel = 'Monitor Brand',figsize=(14, 6)) 
```
